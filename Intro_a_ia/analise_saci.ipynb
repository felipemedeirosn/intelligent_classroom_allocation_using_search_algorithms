```python
import pandas as pd
saci_df = pd.read_csv("tabela_saci.csv")
display (saci_df)
```

```python
#Em quantas turmas cada professor ministra aula?

turmas_professor = saci_df['Professor(a)'].value_counts()
pd.set_option('display.max_rows', None)
display(turmas_professor)
```

```python
#Quantas turmas tem aula em cada uma das salas?

turmas_professor = saci_df['Local'].value_counts()
pd.set_option('display.max_rows', None)
display(turmas_professor)
```
